### Justification 
Time-based interpolation estimates missing values using nearby dates, preserving the natural trend in the data. This approach is more appropriate than using a global mean or median because it maintains the temporal structure of the dataset.

```python
df1['pm25'] = df1['pm25'].interpolate(method='time')
df1
```

### Data quality analysis: city with highest % invalid/missing pm25.


```python
missing_pct = (
    df["pm25"].isna()
        .groupby(df["city"])
        .mean() * 100
)

worst_city = missing_pct.idxmax()
worst_value = missing_pct.max()

worst_city, round(worst_value, 2)
```

This calculates the 7-day rolling average of PM25 separately for each city. For every city, it takes the current day and the previous 6 days and computes their mean. This helps smooth daily fluctuations and makes the overall pollution trend easier to observe.

```python
rolling_analysis = df1.groupby('city')['pm25'].rolling(window=7).mean()
rolling_analysis
```

### Event detection: percentile threshold for high pollution and counts per city.

```python
high_pollution_level = np.percentile(df1['pm25'], 95)  # I selected 95% level for high pollution
high_pollution_level
```

```python
high_pollution_values = df1[df1['pm25'] > high_pollution_level]
high_pollution_values.head()
```

```python
high_pollution_percentage_overall = (len(high_pollution_values) / len(df1))*100    # Event detection: percentile threshold for high pollution
high_pollution_percentage_overall
```

```python
high_pollution_counts = high_pollution_values.groupby('city')['pm25'].count()   # high pollution counts per city
high_pollution_counts
```

### Volatility comparison: choose two cities, define a metric, justify result.

```python
cities = df1[df1['city'].isin(['London', 'Madrid'])].copy()
cities
```

```python
cities.groupby('city')['humidity_pct'].std()  # Standard deviation was used as a measure of volatility because it quantifies how much individual humidity values deviate from the mean. A higher standard deviation indicates greater variability over time. In this case, Madrid shows slightly higher volatility than London, meaning its humidity levels fluctuate more around the average.
```

### Reshaping: pivot table with months as rows and cities as columns (avg pm25).

```python
pivot_pm25 = df1.pivot_table(
    index=df1.index.to_period('M'),
    columns='city',
    values='pm25',
    aggfunc='mean'
)

pivot_pm25
```

2. How does missing or invalid data affect your confidence in the results?

Missing or invalid data can affect averages, hide extreme values, and reduce the overall reliability of the results. Therefore, conclusions may not fully reflect the true pollution levels. In our data, Missing pm25 values reduce confidence in the results because some data points were estimated using interpolation instead of real measurements. Although interpolation helps keep the trend smooth and complete, it may slightly change the true pollution levels, especially extreme values. Therefore, the results should be interpreted with some caution.


3. Does temperature appear related to pm25 in your analysis? Explain briefly.

I used correlation to find the relation between 2 metrics. Yes, temperature appears to be related to PM25 levels. The correlation analysis shows a strong negative relationship (r = -0.72), indicating that as temperature increases, PM25 levels tend to decrease. This suggests an inverse association between temperature and air pollution in the dataset.

4. What is one limitation of using daily averages for air-quality policy decisions?

Daily averages can hide short-term spikes in pollution, which may underestimate actual health risks. They smooth out hourly fluctuations, making it harder to detect dangerous peaks that occur within a single day. As a result, policy decisions based only on daily averages may miss opportunities for timely interventions.
